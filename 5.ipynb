```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 15, figsize=(20,20))#get 15 different images
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, 
    channel_shift_range=10., horizontal_flip=True)
"""here these are the augmentations that we done to the picture taken"""
```

```python
chosen_image = random.choice(os.listdir('...'))#choose a random picture
```

```python
image_path = '...' + chosen_image
```

```python
image = np.expand_dims(plt.imread(image_path),0)#read the dimentions of the image to pass to the augmentation
```

```python
plt.imshow(image[0])
```

```python
aug_iter = gen.flow(image)#give the image data to the generator(augmentation) to change
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(15)]#get different 10 images
```

```python
plotImages(aug_images)
```
